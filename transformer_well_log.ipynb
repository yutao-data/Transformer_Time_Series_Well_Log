## Explorory Analysis

```python
file_path = '/content/drive/MyDrive/CentraleSupelec/Well Log/facies_data.csv'
facies_data = pd.read_csv(file_path)

# Display basic information about the dataset
facies_data_info = facies_data.info()
facies_data_head = facies_data.head()

facies_data_info, facies_data_head
```

Columns Description

1. **Facies**: Discrete rock classes represented as integers.
2. **Formation**: Geological formation names.
3. **Well Name**: Name of the well.
4. **Depth**: Depth in the well.
5. **GR**: Gamma ray measurements.
6. **ILD_log10**: Log of resistivity measurements.
7. **DeltaPHI**: Porosity index difference.
8. **PHIND**: Average of neutron and density logs.
9. **PE**: Photoelectric effect log.
10. **NM_M**: Nonmarine-marine indicator (binary).
11. **RELPOS**: Relative position (geological).

```python
numerical_columns = ["Depth", "GR", "ILD_log10", "DeltaPHI", "PHIND", "PE", "RELPOS"]
stats = facies_data[numerical_columns].describe()
stats
```

```python
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(facies_data[column], kde=True, bins=30, color="blue", alpha=0.7)
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
# Facies distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=facies_data, x="Facies", palette="viridis")
plt.title("Distribution of Facies Classes")
plt.xlabel("Facies")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = facies_data[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
